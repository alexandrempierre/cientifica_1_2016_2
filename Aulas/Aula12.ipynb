# A fórmula de Simpson

Em vez de usarmos uma aproximação "de Riemann" para a integral (ou seja, com apenas um ponto),
vamos usar a idéia principal das outras demonstrações:
queremos estimar com maior precisão as integrais
$$ I _ {n,k} = \int_{c_k}^{d_k = c_k + h} f(x)\,dx. $$

É claro (pense porquê!) que se só tivermos acesso a um ponto, o melhor que podemos fazer é usar o ponto médio $m_k = \frac{c_k + d_k}{2}$.
Se só tivermos dois pontos, a situação é mais complicada.
Usando $c_k$ e $d_k$, que é a fórmula do trapézio, nem fica melhor do que apenas o ponto médio.
Mas podemos combinar os pontos extremos $c_k$ e $d_k$ com o ponto médio, e ter 3 pontos.

Com um pouco de trabalho, podemos mostrar que a combinação certa será:
$$ I _ {n,k} \sim h \, \frac{f(c_k) + 4 f(m_k) + f(d_k)}{6}. $$

```python
def midpoint(f,a,b, n):
    """ Midpoint formula: integrates  f  over the interval  [a,b]  with  n  subdivisions."""
    h = (b-a)/float(n)
    askew = a + h/2
    ints = [f(askew+k*h) for k in range(n)]
    return sum(ints)*h
```

# Métodos gerais para fórmulas

O importante de uma fórmula de integração é obter uma boa aproximação da integral $I_{n,k} = \int_{c_k}^{d_k} f(t) dt$.
Para isso podemos proceder de duas formas equivalentes:

1. Expandindo em série de Taylor (em torno do ponto médio, para simplificar)
2. Formando um sistema de equações para determinar os pesos de cada ponto

## Série de Taylor

Vamos supor que vamos integrar $I = \int_{-1}^1 f(t) dt$, para simplificar as contas.
Expandindo $f$ em série de Taylor no zero, e calculando as integrais, obtemos:
$$ I = 2f(0) + \frac{2}{3} \frac{f''(0)}{2!} + \frac{2}{5} \frac{f''''(0)}{4!} + \cdots
     = \sum_{n = 0}^{\infty} \frac{2}{2n + 1} \frac{f^{(2n)}(0)}{(2n)!} $$

Assim, para obter aproximações cada vez melhores, basta calcular cada vez melhor as derivadas de $f$.
Isso pode ser ou não viável: calcular derivadas simbólicas é (muitas vezes) possível,
mas pode ser mais demorado do que calcular derivadas numéricas,
ainda mais que seremos obrigados a truncar em algum momento.

Por exemplo, podemos aproximar a derivada segunda "iterando" a fórmula das diferenças centradas:
$$ f''(0) \sim f'(1/2) - f'(-1/2) \sim [f(1) - f(0)] - [f(0) - f(-1)] = f(1) - 2f(0) + f(-1). $$

(Observação prática importante: o triângulo de Pascal com sinais alternados calcula derivadas!
A quarta derivada pode ser aproximada por:
$$f''''(0) \sim f(2) - 4f(1) + 6f(0) - 4f(-1) + f(-2).$$

As derivadas de ordem ímpar ficam em pontos de coordenadas "meias", como por exemplo a terceira derivada
$$f'''(0) \sim f(3/2) - 3f(1/2) + 3f(-1/2) - f(-3/2).$$)

Assim, podemos aproximar $I \sim 2f(0) + \frac{2}{3}\frac{f(1) - 2f(0) + f(-1)}{2!} = 2\frac{f(1) + 4f(0) + f(-1)}{6}$,
que é a fórmula de Simpson.

O mais interessante deste método é que ele mostra que o erro será dado pelo "primeiro termo desprezado".
Assim, a fórmula de Simpson terá um erro onde entra a quarta derivada,
e será exata para polinômios de grau 2 (por incorporar a segunda derivada) e 3 ("de bonus", por ser simétrica).

Para maiores ordens, é necessário levar em consideração também que a fórmula de aproximação "binomial"
da derivada discreta tem um erro, dado pela derivada de ordem 2 a mais (de novo por razões de simetria).
Portanto, para obter uma fórmula exata em grau 4,
temos não apenas que aproximar a quarta derivada para "cancelar" o termo $\frac{2}{5} \frac{f''''(0)}{4!}$,
mas também para melhorar a qualidade da aproximação da segunda derivada.

## Equações para os coeficientes.

Podemos mudar o ponto de vista e fazer "ao contrário":
imagine que vamos escolher $n$ pontos à direita do zero, e os seus $n$ simétricos à esquerda do zero.
Sejam $x_i$ os pontos positivos escolhidos.
Uma fórmula para a integral tem a forma:
$$ I = \int _ {-1}^1 f(t) dt \sim 2\gamma f(0) + \sum \alpha_i [f(x_i) + f(-x_i)]. $$

Agora, em vez de expandir tudo em série de Taylor,
vamos pedir que as integrais sejam exatas para polinômios de grau baixo.
Por simetria, as equações com polinômios de grau ímpar são identicamente $0 = 0$.
Por outro lado, restam as equações de grau par:
$$ \frac{2}{2k + 1} = 2\gamma 0^{2k} + \sum \alpha_i [x_i^{2k} + (-x_i)^{2k}]
  = 2\left[\gamma 0^{2k} + \sum \alpha_i x_i^{2k}\right]. $$

Temos portanto $n+1$ incógnitas (os pesos $\gamma$ e $\alpha_i$)
e portanto podemos obter igualdade em $n+1$ equações, indo portanto até grau $2n$, e o erro será de ordem $2n+2$.
Em forma matricial, simplificando o fator comum $2$,
e escrevendo $y_i = x_i^2$ para simplificar, temos:
$$
\begin{bmatrix} 1 \\ 1/3 \\ 1/5 \\ 1/7 \\ 1/9 \\ \end{bmatrix} =
\begin{bmatrix}
1 & 1     & 1     & 1     & 1 \\
0 & y_1   & y_2   & y_3   & y_4   \\
0 & y_1^2 & y_2^2 & y_3^2 & y_4^2 \\
0 & y_1^3 & y_2^3 & y_3^3 & y_4^3 \\
0 & y_1^4 & y_2^4 & y_3^4 & y_4^4 \\
\end{bmatrix}
\begin{bmatrix} \gamma \\ \alpha_1 \\ \alpha_2 \\ \alpha_3 \\ \alpha_4 \\ \end{bmatrix}
$$

Esta matrix é conhecida como "matriz de Vandermonde",
e contanto que os $y_i$ sejam diferentes entre si
(o que é totalmente razoável nas nossas condições!),
ela é inversível.
Ou seja, é realmente possível resolver este sistema e obter uma solução.


Este método tem a vantagem de ser mais fácil ainda obter os coeficientes do que a expansão via série de Taylor,
pois basta resolver um sistema linear.
Além disso, ele é mais geral, pois poderíamos também alterar os $x_i$ (em vez de tomar pontos equiespaçados)
o que nos dá a possibilidade de "ter mais incógnias" (os $x_i$!) e assim satisfazer mais equações ainda.
As fórmulas de integração que usam esta técnica são conhecidas como "de Gauss-Legendre".

# Fórmulas adaptativas

O problema maior da integração é que, ao diminuir $h$, aumentamos o tempo de cálculo.
O que gostaríamos de fazer é, então, poder variar $h$ ao longo do algoritmo,
em função do intervalo que estejamos calculando.

Ou seja, vamos ter que achar, dependendo da função, os valores de $h$ "certos".
Mais ainda, gostaríamos que o método fosse "de passo variável":
isso permite "dar um zoom" quando a função variar muito (segunda ou quarta derivada de $f$,
dependendo de usar ponto médio ou Simpson)

## Recursão, o retorno

Uma solução para este problema é recursiva

1. Tomamos um intervalo $I$ sobre o qual vamos calcular uma "boa" aproximação da integral
2. Usamos alguma fórmula para calcular sua aproximação $A$
3. Usamos outra fórmula, teóricamente mais precisa, para calcular uma segunda aproximação $A'$.
4. Com $A$ e $A'$, podemos estimar o **erro** cometido pelas integrais
    1. Se ambas aproximações estiverem próximas, o erro deve ser pequeno, e usamos $A'$.
    2. Senão, dividimos o intervalo em duas partes (esquerda e direita, olhaí a bisseção) $I_e$ e $I_d$,
       calculamos ambas as integrais com precisão suficiente, e usamos a soma.

```python
def simpson_adaptativo(f, a,b, tol=1e-6):
    h = (b-a)
    m = a + h/2

    fa = f(a)
    fb = f(b)
    fm = f(m)
    
    # II e III
    Igrossa = h*(fa + 4*fm + fb)/6
    Ifina   = h*(fa + 4*f((a+m)/2) + 2*fm + 4*f((m+b)/2) + fb)/12
    
    # IV
    if abs(Igrossa - Ifina) < tol: return Ifina
    else:
        esquerda = simpson_adaptativo(f, a,m, tol/2)
        direita  = simpson_adaptativo(f, m,b, tol/2)
        return esquerda+direita
```

```python
def simpson_adaptativo(f, a,b, tol=1e-6):
    
    def loop(a,b,m,fa,fb,fm,Icur,tol):
        h = (b-a)
        m1 = a + h/4
        m2 = b - h/4
        fm1 = f(m1)
        fm2 = f(m2)
    
        # II e III (II = Iprev)
        Ileft  = h*(fa + 4*fm1 + fm)/12
        Iright = h*(fm + 4*fm2 + fb)/12
        Inext = Ileft + Iright
    
        # IV
        if abs(Icur - Inext) < 15*tol/2: return Inext, 2
        else:
            esquerda, ne = loop(a,m,m1,fa,fm,fm1, Ileft,  tol/2)
            direita,  nd = loop(m,b,m2,fm,fb,fm2, Iright, tol/2)
            return esquerda+direita, ne + nd + 2
    
    m = (a+b)/2
    h = (b-a)
    fa = f(a)
    fb = f(b)
    fm = f(m)
    I0 = h*(fa + 4*fm + fb)/6
    I, niter = loop(a,b,m,fa,fb,fm, I0,tol)
    
    return I,niter+3
```
