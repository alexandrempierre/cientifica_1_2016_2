## `Plot` = Sobrevivência na selva

O comando mais importante para fazer gráficos é, sem dúvida, `plot`.
Ele tem diversas opções (veja a ajuda correspondente), mas em geral as configurações-padrão são bastante razoáveis.
Vejamos alguns exemplos.

```python
# Plotar uma lista, usando o índice como coordenada $x$

lista_random = rand(55) # rand vem do numpy
plot(lista_random);
```

```python
# A segunda forma mais útil é plotar uma função.
x = range(100)
y = sin(x) # Esse sin() vem do numpy também. O sin() da math não sabe calcular num `range`
plot(x, y);
```

```python
# A terceira, variação da anterior, é uma curva paramétrica
t = arange(0,10,0.1)
xt = t**2 * sin(t)
yt = t*cos(t)
plot(xt,yt);
```

Neste exemplo, usamos `arange`, que é uma função do _NumPy_, que retorna um "vetor numérico".
A grande vantagem destes vetores com relação às listas é que é possível fazer operações numéricas (coordenada-a-coordenada)
com mais facilidade e clareza do que com uma lista "normal" do Python.
Isso será bastante útil durante o curso, seja para gerar dados, seja para escrever programas mais simples.

E a `matplotlib` usa vetores do NumPy automaticamente.

### Exercício resolvido:

Fazer um gráfico com os dois pontos extremos dos intervalos do método da bisseção.

Vamos adaptar o método da bisseção para obter estas informações suplementares.

```python
def bissecao_dados(f,a,b,prec=1e-6):
    """ Método da bisseção para uma função f no intervalo [a,b], retornando a seqüência dos pontos extremos. """
    m = (a+b)/2
    # Se já há precisão suficiente, retornamos o ponto médio e o último intervalo
    if abs(b - a)/2 < prec: return m, [a],[b]

    # Se f(m) == 0, achamos uma raiz exata!
    if f(m) == 0: return m, [a],[b]

    # Senão, vamos por recorrência
    if f(m)*f(a) < 0:
        z, esq, dire = bissecao_dados(f,a,m,prec)
        return z, [a]+esq, [b]+dire
    else:
        z, esq, dire = bissecao_dados(f,m,b,prec)
        return z, [a]+esq, [b]+dire
```

```python
def plot_bissecao(z, esquerda, direita):
    esquerda = array(esquerda)
    direita = array(direita)
    plot(esquerda, label='liminf')
    plot(direita, label='limsup')
    legend()
    figure(figsize=(12,4))
    subplot(121)
    semilogy(direita - esquerda)
    title('Tamanho do intervalo = erro absoluto máximo')
    subplot(122)
    semilogy(direita - z, label='limsup')
    semilogy(z - esquerda, label='liminf')
    legend()
    title('Distância entre cada extremo e a raiz real')
```

```python
f = cos
a = 0.
b = 3.

z, esq, dire = bissecao_dados(f, a, b)
print("Uma raiz de f entre {} e {} é {}.".format(a, b, z))
plot_bissecao(z, esq, dire)
```

### Exercício

Agora, adapte estas funções para fazer um gráfico do andamento do método de Newton.

```python
def newton_dados(f,df,x,tol=1e-8,reltol=1e-8,niter=100):
    dx = f(x)/df(x)
    eabs = abs(dx)
    if eabs < tol or eabs/abs(x) < reltol: return x, [x]
    
    z,zs = newton_dados(f,df,x-dx, tol,reltol,niter-1)
    return z, [x]+zs
```

```python
def plot_newton(z,x):
    x = array(x)
    figure(figsize=(12,4))
    subplot(121)
    plot(x, 'bo')
    subplot(122)
    semilogy(abs(x - z))
    title('Distância entre cada chute e a raiz real')
```

```python
def dcos(x):
    return -sin(x)

z,zs = newton_dados(cos,dcos,4, tol=1e-12, reltol=1e-12)
plot_newton(z,zs)
```

```python
z,e,d = bissecao_dados(cos,3,5,prec=1e-12)
plot_bissecao(z,e,d)
```

```python
def p3(x):
    return x**3 - 4*x + 7
def dp3(x):
    return 3*x**2 - 4
```

```python
plot_newton(*newton_dados(p3,dp3,1.84))
```

# Complexidade, exemplos

Lembremos da função fatorial

```python
def fatorial(n):
    assert (isinstance(n, int))
    assert (n >= 0)
    
    p = 1
    while(n > 1):
        p *= n
        n -= 1

    return p
```

Vejamos quanto tempo leva para calcular alguns fatoriais.
Para isso, utilisamos o módulo `timeit`.

```python
import timeit
```

```python
# No ipython, temos o %timeit, um "comando mágico":
%timeit(fatorial(10))
%timeit(fatorial(100))
%timeit(fatorial(1000))
%timeit(fatorial(10000))
```

```python
# Senão, podemos usar o comando Python "puro"
timeit.timeit(number=100, setup="from __main__ import fatorial", stmt="x = fatorial(1000)")
```

## Dois códigos, duas performances

```python
def binom(n,k):
    assert(isinstance(n,int) and isinstance(k,int))
    assert(0 <= n and 0 <= k and k <= n)
    
    i = 0
    p = 1
    while(i < k):
        p = (p * (n-i))/(i+1)
        i += 1
    return p
```

```python
def slow_binom(n,k):
    assert(isinstance(n,int) and isinstance(k,int))
    assert(0 <= n and 0 <= k and k <= n)
    
    return fatorial(n)/fatorial(k)/fatorial(n-k)
```

```python
print("20 escolhe 10")
%timeit(binom(20,10))
%timeit(slow_binom(20,10))

print("200 escolhe 10")
%timeit(binom(200,10))
%timeit(slow_binom(200,10))

print("200 escolhe 100")
%timeit(binom(200,100))
%timeit(slow_binom(200,100))
```

```python
print("Rápida")
%timeit(binom(20,10))
%timeit(binom(200,10))
%timeit(binom(2000,10))
%timeit(binom(20000,10))

print("\nDevagar")
%timeit(slow_binom(20,10))
%timeit(slow_binom(200,10))
%timeit(slow_binom(2000,10))
%timeit(slow_binom(20000,10))
```
