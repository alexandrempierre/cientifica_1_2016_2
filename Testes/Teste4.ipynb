# Erros de aproximação _vs_ erros numéricos

Possuímos duas fórmulas para aproximar a derivada de uma função $f$: a derivada lateral e a derivada central.
O objetivo deste teste é estudar quando o erro numérico fica mais importante do que o erro de aproximação.

## Parte 1: A derivada central

```python
def d_lateral(f,x,eps):
    return (f(x+eps) - f(x))/eps
```

### Questão 1

Faça um gráfico da derivada lateral numérica, no intervalo $[1,2]$, variando `eps` entre `1e-16` e `1e-4`.
Use as seguintes funções:

- sin(x)
- exp(x)
- log(x)

```python
eps = 10.**arange(-16,-3,2)
x = arange(1,2+1e-3,1e-3)
y = [-1]*len(x) # y=-1 para que a reta y=0 seja exibida

eps_print = { 1e-16:'10^{-16}', 1e-14:'10^{-14}', 1e-12:'10^{-12}', 1e-10:'10^{-10}', 1e-8:'10^{-8}', 
            1e-6:'10^{-6}', 1e-4:'10^{-4}'}

def plot_df(f, df_num=d_lateral, titulo='', local_x=x, local_eps=eps):
    title(titulo)
    for e in local_eps:
        plot(local_x,df_num(f,local_x,e), label='$\epsilon={}$'.format(eps_print[e]))
        legend(loc=0)
```

```python
figure(figsize=(18,6))

subplot(131)
plot_df(sin, titulo='$\sin(x)$')

subplot(132)
plot_df(exp, titulo='$e^x$')
plot(x,y)

subplot(133)
plot_df(log, titulo='$\log(x)$')
plot(x,y)

pass
```

### Questão 2

Quais são os valores que a derivada apresenta com $\epsilon = 10^{-16}$? Porque isso faz sentido?

Para $\epsilon = 10^{-16}$, o valor da derivada numérica lateral está sendo zero. 

Isso acontece por causa da representação em ponto flutuante, o erro relativo do intervalo é o limite para o quão pequenos podem ser os números representados. Aparentemente entre $0$ e $1$ esse erro relativo é maior que $10^{-16}$.

### Questão 3

Para entender melhor o que aconteceu, faça um gráfico (em escala logarítmica) do erro.

Dica: escreva uma _função_ "plota_erro" que recebe a função que você vai plotar e a derivada, e faz o resto sozinha.
Depois, chame esta função três vezes.
Se quiser, dê alguns argumentos a mais para customizar os gráficos.

```python
def plota_erro(f, df, df_num=d_lateral, titulo='', local_x=x, local_eps=eps):
    title(titulo)
    for e in local_eps:
        y = df(x) - df_num(f,local_x,e)
        semilogy(x, y, label='$\epsilon={}$'.format(eps_print[e]))
        legend(loc=0)
```

```python
figure(figsize=(18,6))

subplot(131)
plota_erro(sin, cos, titulo='erro de $\sin(x)$')

subplot(132)
plota_erro(exp, exp, titulo='erro de $e^x$')

subplot(133)
plota_erro(log, lambda x: 1/x, titulo='erro de $\log(x)$')
```

### Questão 4: o melhor $\epsilon$

Queremos ver onde deve estar o melhor valor de $\epsilon$ para $x \in [1,2]$.
Vamos nos concentrar na função exponencial.

Faça novos gráficos / contas para buscar o $\epsilon$ que dá o "menor erro".

```python
# Faça aqui (e abaixo, se quiser) contas, gráficos, etc

def intervalo(f, df, df_num=d_lateral, local_x=x, local_eps=eps):
    #max_abs = lambda l: max(abs(l))
    media = lambda l: abs(sum(l)/len(l))
    
    menor_erro, epsilon = 1e+100, 1e+100
    
    for e in local_eps:
        #erro = max_abs(df(x) - df_num(f, x, e))
        erro = media(df(x) - df_num(f, x, e))
        if erro < menor_erro:
            menor_erro, epsilon = erro, e
    
    r = ''
    i = list(local_eps).index(epsilon)
    
    if epsilon == local_eps[0]:
        r = '{} <= epsilon < {}'.format(epsilon, local_eps[i+1])
    elif epsilon == local_eps[-1]:
        r = '{} < epsilon <= {}'.format(local_eps[i-1], epsilon)
    else:
        r = '{} < epsilon < {}'.format(local_eps[i-1], local_eps[i+1])
    
    return r

intervalo(exp, exp)
```

Agora, interprete os gráficos e dê o melhor intervalo que você pode para onde deve estar $\epsilon$.

Dentre os valores verificados o melhor é o $10^{-8}$.

Mas não tenho como dizer que $\not \exists \epsilon : 10^{-10} \lt \epsilon \lt 10^{-6}, \epsilon \neq 10^{-8}$ que induz um erro menor.

Então, concluo que $10^{-10} \lt \epsilon \lt 10^{-6}$.

## Parte 2: Diferenças centradas

Vamos repetir o que foi feito no exercício anterior, agora para o novo método numérico.
É aqui que ter feito as funções no exercício anterior vai ajudar muito!

```python
def d_central(f,x,eps):
    return (f(x+eps) - f(x-eps))/(2*eps)
```

### Questão 1: Gráficos da derivada

```python
figure(figsize=(18,6))

subplot(131)
plot_df(sin, d_central, '$\sin(x)$')

subplot(132)
plot_df(exp, d_central, '$e^x$')
plot(x,y)

subplot(133)
plot_df(log, d_central, '$\log(x)$')

pass
```

### Questão 2

Você nota alguma diferença significativa entre os gráficos? Porque você acha que isso acontece?

A diferença entre as derivadas numéricas centrais para $\epsilon = 10^{-14}$ e para expoentes menores parece ter diminuído e há um segmento vertical provavelmente associado ao primeiro valor de $x$.

O primeiro valor de $x$ é $1$ que está na fronteira de intervalos, um em que o erro relativo zera qualquer $n : \lvert n \rvert \leq 10^{-16}$ e outro com um erro relativo de módulo um pouco menor. A derivada central usa o resultado da função para dois valores: um que acaba sendo menor que $1$, e outro que por causa do erro acaba igual a $1$.

Como o resultado das funções $\sin(x)$ e $\log(x)$ é menor ou igual a $1$ nos valores avaliados, esse valor não anulado é plotado.

Se nenhum parâmetro for passado impedindo isso, a _matplotlib_ tenta fazer um gráfico contínuo. Por isso o segmento vertical, na verdade há uma quantidade muito pequena (provavelmente um único) de valores da derivada numérica com $\epsilon = 10^{-16}$ que não dá $0$ mesmo para $\sin(x)$ e $\log(x)$.

No caso de $\epsilon = 10^{-14}$, vale uma argumentação parecida, mas o $\epsilon$ em si não é anulado, então o que ocorre na derivada central é que o ruído no gráfico apenas diminui.
