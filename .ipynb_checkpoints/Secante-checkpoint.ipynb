```python
def bissecao(f,a,b,prec=1e-10):
    m = (a+b)/2
    
    if f(m) == 0: return m
    if abs(b-a)/2 < prec: return m
    
    if f(a)*f(m) < 0: return bissecao(f,a,m)
    else: return bissecao(f,m,b)
```

```python
def reta_secante(f,x0,x1):
    incl = (f(x1) - f(x0))/(x1 - x0)
    base = f(x0)
    
    def reta(x):
        return incl*(x-x0)+base
    
    return reta

def plota_secantes(f,x0,x1):
    q_alvos = 5
    q_pontos = 1e+3
    
    alvos = [x0+i*(x1-x0)/q_alvos for i in range(1,q_alvos+1)]
    secantes = [(alvo, reta_secante(f,x0,alvo)) for alvo in alvos]
    
    figure(figsize=(14,6))
    
    x = arange(-x1, x1, (x1-x0)*1e-3)
    plot(x, f(x), label='f(x)')
    
    for (alvo, s) in secantes:
        plot(x, s(x), '--', label='secante 0 -- {:.2}'.format(alvo))
        
    legend(loc=0)

plota_secantes(cos, 0, 2*pi)
```

```python
def secante(f,x0,x1,prec=1e-10):
    #if abs(f(x1)-f(x0)) == 0:
    #    x1 = f(x1)
        
    x2 = x0 - f(x0)*(x1-x0)/(f(x1)-f(x0))
    
    if abs(x2-x1) < prec or f(x2) == 0: return x2
    
    return secante(f,x1,x2)
    
secante(sin, pi/4, pi/2)
```

```python
f = lambda x: x**2

x0, x1 = -2, 2

bissecao(f,-2,2), secante(f,x0,x1 if abs(f(x1)-f(x0)) != 0 else x1+10e-10)
```

```python
%timeit bissecao(f,0,2)
```
